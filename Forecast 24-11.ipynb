## Importar archivos, concatenar, eliminar columnas no necesarias, agrupar por material y fecha. Se muestran solo los 20 primeros elementos por capacidad del programa.

```python
import pandas as pd
import os
# lista con los archivos
file_names = ["SLM2019 .xlsx", "SLM2020.xlsx", "SLM2021.xlsx", "SLM2022.xlsx", "SLM2023.xlsx"]
# almacenar los df
dfs = []

for file_name in file_names:
    xls = pd.ExcelFile(file_name)
    for sheet_name in xls.sheet_names:
        df = xls.parse(sheet_name)
        df['File'] = file_name  # identificar de que archivo viene
        df['Sheet'] = sheet_name  #identificar de que hoja proviene
        dfs.append(df)

# concatenar todos los dfs en uno solo
combined_df = pd.concat(dfs, ignore_index=True)
# eliminar nulos en 'Material Description'
combined_df = combined_df.dropna(subset=['Material Description'])

# 'mes-año' a tipo de dato datetime
combined_df['mes-año'] = pd.to_datetime(combined_df['mes-año'], errors='coerce')

# agrupar por 'Material Description' y 'mes-año'
grouped_df = combined_df.groupby(['Material Description', 'mes-año']).first().reset_index()
# eliminar las columnas que se me pasaron de los excels
columns_to_drop = ['Cons.120/36 months', 'Ddisponent', 'Plant', 'Valuation Class', 'Base Unit of Measure', 'Stock market price(LPP)']
grouped_df = grouped_df.drop(columns=columns_to_drop)
# guardar df en un excel (son muy grandes si)
grouped_df.to_excel("combined_data.xlsx", index=False)

grouped_df.head(20)
```

### Como hay materiales que se encontraban en el 2019 pero los años siguientes salieron de bodega, o no se encontraban en los primeros años pero si en los finales, cada vez que estos elementos no aparecían en un año, se reemplaza el valor nulo por un 0 para poder trabajarlos.

### Se ve la nueva tabla.
